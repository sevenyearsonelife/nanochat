```python
import torch
import torch.nn as nn
import torch.nn.functional as F



# 一个简单模型
class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(784, 256)
        self.fc2 = nn.Linear(256, 10)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        return self.fc2(x)

model = Net()

x = torch.randn(1, 784)
y = torch.randint(0, 10, (1,))

out = model(x)

#print(f"x: {x}")
print(f"y: {y}")
print(f"out: {out}")

criterion = nn.CrossEntropyLoss()


loss = criterion(out, y)
print(f"loss: {loss}")
```
